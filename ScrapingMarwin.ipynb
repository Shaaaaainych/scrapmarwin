```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
url = "https://www.marwin.kz/press/top-50-sovremennoi-prozi.html?from=knigisis&product_list_mode=list"

r = requests.get(url)
r.text
```

```python
soup = BeautifulSoup(r.text, 'lxml')
soup
```

```python
title = soup.find('div', class_ = 'product-item-info').find('a', class_ = 'product-item-link').text
title
```

```python
price = soup.find('span', class_ = 'price').text
price
```

```python
age = "Возраст: " + soup.find('td', class_ = 'col data').text
age
```

```python
company = "Издательство: " + soup.findAll('td', class_ = 'col data')[2].text
company
```

```python
data = []

for p in range(1,6):
    
    url = "https://www.marwin.kz/press/top-50-sovremennoi-prozi.html/?from=knigisis&p={p}&product_list_mode=list"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    sleep(3)
    
    books = soup.findAll('div', class_ = 'product-item-info')

    for book in books:
        title = book.find('a', class_ = 'product-item-link').text
        price = book.find('span', class_ = 'price').text
        age = "Возраст: " + soup.find('td', class_ = 'col data').text
        company = "Издательство: " + soup.findAll('td', class_ = 'col data')[2].text
        data.append([title, price,age, company])
```

```python
header = ['title', 'price','age', 'company']
df = pd.DataFrame(data, columns = header)
df.to_csv('/Users/Shaaaaaaaain/Desktop/marwin_data.csv', sep = ';', encoding = 'utf8')
```
